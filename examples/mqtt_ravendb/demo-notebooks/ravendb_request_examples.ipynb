```python
import json
import requests
```

## Get Production Order Header by Order Number

```python
query = """
        from 'productionorders' 
        where search(payload.orderNumber, '000001320749')
        """

response = requests.post(
    url="http://wiengdevdbsrv.williams-int.com/databases/Mcp/queries",
    json={"Query": query},
    headers={"Content-Type": "application/json"}
)

if response.status_code == 200:
    result = response.json()
    documents = result.get('Results', [])

    for doc in documents:
        pretty_doc = json.dumps(doc, indent=4)
        print(pretty_doc)
else:
    print(f"Error {response.status_code}: {response.text}")
```

## Get Count of Production Orders in Cell

```python
query = """
        from 'productionorders'
        group by payload.cell
        select payload.cell as Cell, count() as Count
        """

response = requests.post(
    url="http://localhost:8080/databases/Mcp/queries",
    json={"Query": query},
    headers={"Content-Type": "application/json"}
)

if response.status_code == 200:
    result = response.json()
    documents = result.get('Results', [])

    for doc in documents:
        pretty_doc = json.dumps(doc, indent=4)
        print(pretty_doc)
else:
    print(f"Error {response.status_code}: {response.text}")
```
